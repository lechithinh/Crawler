```python
from facebook_scraper import get_posts
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
FANPAGE_LINK ="UITconfess"
FOLDER_PATH = "/content/drive/MyDrive/Projects/Crawler"
COOKIE_PATH = "/content/drive/MyDrive/Projects/Crawler/cookies.txt" #should generate your own cookies in facebook

PAGES_NUMBER = 5 
```

```python
post_list = []
for post in get_posts(FANPAGE_LINK, 
                    options={"comments": True, "reactions": True, "allow_extra_requests": True}, 
                    extra_info=True, pages=5, cookies=COOKIE_PATH):
    post_list.append(post)
```

```python
len(post_list)
```

```python
post_list[0]
```

```python
fields_list = ['post_id', 'text', 'time', 'images', 'likes', 'comments', 'shares', 'post_url', "NumOfComments"]
result_list = []
for post in post_list:
    record = []
    for field in fields_list[:-1]:
         record.append(post[field])
    record.append(len(post["comments_full"]))
    result_list.append(record)
```

```python
post_df = pd.DataFrame(result_list, columns = fields_list)
post_df
```

```python
post_df.to_csv(FANPAGE_LINK, encoding='utf-8', index=False)
```

```python
attributes = ["comment_id",'commenter_name', 'comment_text', 'commenter_url', 'comment_time', 'comment_reaction_count']
for post in post_list:
    id_name = post["post_id"]
    if len(post['comments_full']) > 0:
      record = []
      print(f"post: {id_name} has {len(post['comments_full'])}")
      for cmt in  post['comments_full']:
          each_cmt = []
          for attr in attributes:
            each_cmt.append(cmt[attr])
          record.append(each_cmt)
      df = pd.DataFrame(record, columns = attributes)
      df.to_csv(id_name, encoding='utf-8', index=False)
    else:
      print(f"{id_name} has no comments")
    
```
